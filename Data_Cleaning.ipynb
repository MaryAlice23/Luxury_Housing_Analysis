```python
# Importing Important Libraries
import numpy as np
import pandas as pd
```

```python
# Importing my Dataset
df=pd.read_csv("Path_of_Dataset")
```

```python
# Checking the NULL values in the dataset
df.isnull().sum().any()
```

```python
df.info()
```

```python
# Count NULLs per column
df.isnull().sum().sort_values(ascending=False)
```

```python
df.info()
df.isnull().sum()
```

```python
# Dataset shape
print("Shape:", df.shape)

# Column names
print("\nColumns:\n", df.columns)

# Data types
print("\nData Types:\n")
df.info()
```

```python
# Missing values count
missing_values = df.isnull().sum()
missing_values[missing_values > 0]
```

```python
# Fill numeric columns with median
numeric_cols = df.select_dtypes(include=["int64", "float64"]).columns
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].median())

# Fill categorical columns with mode
categorical_cols = df.select_dtypes(include=["object"]).columns
for col in categorical_cols:
    df[col] = df[col].fillna(df[col].mode()[0])
```

```python
# Check duplicates
print("Duplicate rows:", df.duplicated().sum())

# Drop duplicates
df = df.drop_duplicates()
```

```python
# Strip spaces and standardize text
for col in categorical_cols:
    df[col] = df[col].str.strip().str.title()
```

```python
# Convert Ticket Price to numeric if needed
if "Ticket_Price_Cr" in df.columns:
    df["Ticket_Price_Cr"] = pd.to_numeric(df["Ticket_Price_Cr"], errors="coerce")

# Convert Amenity Score to numeric
if "Amenity_Score" in df.columns:
    df["Amenity_Score"] = pd.to_numeric(df["Amenity_Score"], errors="coerce")
```

```python
# Check unique booking status values
df["Possession_Status"].value_counts()
```

```python
# Check unique booking status values
 
df["Possession_Status"] = df["Possession_Status"].replace({
    "Booked ": "Booked",
    "Not Booked": "Not Booked"
})
```

```python
# Create Quarter column if Date exists
if "Booking_Date" in df.columns:
    df["Booking_Date"] = pd.to_datetime(df["Booking_Date"], errors="coerce")
    df["Quarter"] = df["Booking_Date"].dt.to_period("Q").astype(str)
```

```python
# Final check
print("Final Shape:", df.shape)
print("\nMissing values after cleaning:\n", df.isnull().sum().sum())
```

```python
# Save cleaned dataset
df.to_csv("cleaned_data.csv", index=False)

print("Cleaned dataset saved successfully!")
```
